```python
import requests
import pandas as pd
import time

API_KEY = 'AV4U2ZRJMFOO6XEY'
EMPRESAS = ['NFLX','AAPL','GOOGL','AMZN','MSFT']
lista_emp = []

for simbolo in EMPRESAS:
  print(f'Extraindo dados de {simbolo}')
  time.sleep(1)
  url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={simbolo}&apikey={API_KEY}'
  r = requests.get(url)
  base = r.json()

  if 'Global Quote' in base:
    temporario = pd.DataFrame([base['Global Quote']])
    lista_emp.append(temporario)


base_empresas = pd.concat(lista_emp)
```
